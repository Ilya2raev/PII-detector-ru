### PII Types
Seven Labels to classify from the following types of PII:

- `NAME` - First or second name in the text.
-`EMAIL` - Email address.
-`USERNAME` - Username on any platform.
-`CARD_NUMBER` - 16 digit card number.
-`PHONE_NUM` - A phone number associated with a person.
-`URL_PERSONAL` - A URL that might be used for identification.
-`STREET_ADDRESS` - Address

```python
df = pd.concat([df_0, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_12, df_42], axis=0)
df.reset_index(drop=True, inplace=True)
```

```python
email_regex = r'[\w.+-]+@[\w-]+\.[\w.-]+'
phone_num_regex = r"(\(\d{3}\)\d{3}\-\d{4}\w*|\d{3}\.\d{3}\.\d{4})\s"

def preprocess(df):
    df['labels'] = df['labels'].apply(lambda x: ast.literal_eval(x))
    df['tokens'] = df['tokens'].apply(lambda x: ast.literal_eval(x))
    df['token_len'] = df['tokens'].apply(len)
    df['label_freq'] = df['labels'].apply(Counter)
    df['label_count'] = df['label_freq'].apply(lambda x: len(dict(x)))
    
    # Eliminating punctuation and unnecessary tokens from positive labels
    for i in range(len(df)):
        for idx, tokens in enumerate(df.iloc[i]['tokens']):
            for tok in tokens:
                if tok in 'кпгсу':
#                 if tok in 'кпгсу' and re.fullmatch(email_regex, tok) is None and re.fullmatch(phone_num_regex, tok) is None:
                    df.iloc[i]['labels'][idx] = 'O'  
    
    return df

df = preprocess(df)
```

```python
# Eliminating punctuation and unnecessary tokens from positive labels
for i in range(len(df)):
    for idx, tokens in enumerate(df.iloc[i]['tokens']):
        for tok in tokens:
            if tok in string.punctuation + 'кпгсуд':
                df.iloc[i]['labels'][idx] = 'O'
```
